```python
# Reload the original DataFrame to ensure all columns are present
url = 'https://data.cityofnewyork.us/resource/erm2-nwe9.json?$limit=50000'
df = pd.read_json(url)

# Standardize complaint types
df['complaint_type'] = df['complaint_type'].replace("HEAT/HOT WATER", "HEATING") \
    .replace("Plumbing", "PLUMBING") \
    .replace(("General", "GENERAL", "CONSTRUCTION"), "GENERAL CONSTRUCTION") \
    .replace("Unsanitary Condition", "UNSANITARY CONDITION") \
    .replace("Safety", "SAFETY") \
    .replace("Outside Building", "OUTSIDE BUILDING") \
    .replace("Appliance", "APPLIANCE") \
    .replace("Electric", "ELECTRIC") \
    .replace("PAINT - PLASTER", "PAINT/PLASTER") \
    .replace(("STRUCTURAL", "AGENCY", "VACANT APARTMENT", "Mold"), "OTHER")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import pandas as pd

df['is_illegal_parking'] = (df['complaint_type'] == 'Illegal Parking').astype(int)

features = ['borough', 'incident_zip', 'facility_type', 'vehicle_type']

print("Shape before dropping NaNs:", df.shape)
print("NaN counts in features and target:")
print(df[features + ['is_illegal_parking']].isnull().sum())

df_model = df.dropna(subset=features + ['is_illegal_parking'])

print("Shape after dropping NaNs:", df_model.shape)

for col in features:
    if col in df_model.columns:
        le = LabelEncoder()
        df_model.loc[:, col] = le.fit_transform(df_model[col])
    else:
        print(f"Warning: Column '{col}' not found in df_model after dropping NaNs.")

X = df_model[features]
y = df_model['is_illegal_parking']

if not X.empty and not y.empty:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    model = RandomForestClassifier(random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred))
else:
    print("Cannot perform train_test_split because X or y is empty after dropping NaNs.")
```

```python

df['created_date'] = pd.to_datetime(df['created_date'])

target_complaint = 'Illegal Parking'
df_complaint = df[df['complaint_type'] == target_complaint].copy()

df_complaint['hour'] = df_complaint['created_date'].dt.hour
df_complaint['dayofweek'] = df_complaint['created_date'].dt.day_name()
df_complaint['month'] = df_complaint['created_date'].dt.month_name()

# 1. Complaints by hour of day
plt.figure(figsize=(10, 4))
df_complaint['hour'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title(f'{target_complaint} Complaints by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 2. Complaints by day of week
plt.figure(figsize=(10, 4))
df_complaint['dayofweek'].value_counts().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
).plot(kind='bar', color='blue')
plt.title(f'{target_complaint} Complaints by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Complaints by month
plt.figure(figsize=(10, 4))
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
df_complaint['month'].value_counts().reindex(month_order).plot(kind='bar', color='lightgreen')
plt.title(f'{target_complaint} Complaints by Month')
plt.xlabel('Month')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
